1-Calculate the correlation matrix, which represents the correlation between each pair of variables (columns) in the Androidapp dataset, starting from column 8.

```python
import pandas as pd
```

```python
df= pd.read_csv("E:/downld/Android_malware.csv", low_memory=False)
```

```python
# Calculating the correlation matrix starting from column 8
# Select columns starting from the 8th column (index 7)
columns_to_analyze = df.iloc[:, 7:]

# Compute the correlation matrix
correlation_matrix = columns_to_analyze.corr()
```

```python
correlation_matrix
```

2- Identify the pairs that have a correlation of 1 (100% correlated). It is necessary to consider that the correlation matrix includes redundant pairs, and the lower triangular pairs are repetitions of the pairs in the upper triangle. To remove redundant pairs, utilize the "get_redundant_pairs" function provided in the next cell.

```python
def get_redundant_pairs(df):
    '''Get diagonal and lower triangular pairs of correlation matrix'''
    pairs_to_drop = set()
    cols = df.columns
    for i in range(0, df.shape[1]):
        for j in range(0, i+1):
            pairs_to_drop.add((cols[i], cols[j]))
    return pairs_to_drop
```

```python
# Get redundant pairs
redundant_pairs = get_redundant_pairs(correlation_matrix)

# Find all correlations and remove redundant pairs
all_correlations = correlation_matrix.unstack()

# Drop redundant pairs
unique_correlations = all_correlations.drop(labels=redundant_pairs)
```

3- Print the pair of variables that are 100% correlated (correlation=1)

```python
import numpy as np

# Print columns where correlation is 1.0
print("Columns with 1.0 correlation:")
for index, value in unique_correlations.items():
    if np.isclose(value, 1.0):
        print(index)
```

```python
correlations = correlation_matrix.fillna(0).stack().reset_index()
correlations.columns = ['feature 1', 'feature 2', 'Correlation']
correlations
```

```python
#dropping all the columnets with correlation that is not 1 and 
#dropping one feature from each pair of high correlation
correlations.drop(correlations.loc[correlations['Correlation']!=1].index, inplace=True)
```

```python
#removing the pair
correlations_dropped=correlations.drop('feature 2', axis=1)
correlations_dropped
```

8- Perform PCA and identify the features that contribute to cumulative variance equal to or above 95%.

```python
import pandas as pd
import numpy as np
from sklearn.decomposition import PCA
from sklearn.preprocessing import StandardScaler

X = app_df.iloc[:, 8:]

y = app_df['malware']

scaler = StandardScaler()
scaled_features = scaler.fit_transform(X)

pca = PCA()
pca.fit(scaled_features)

explained_variance_ratio = pca.explained_variance_ratio_

cumulative_variance = np.cumsum(explained_variance_ratio)

n_components = np.argmax(cumulative_variance >= 0.95) + 1

pca = PCA(n_components=n_components)
principal_components = pca.fit_transform(scaled_features)

df_pca = pd.DataFrame(data=principal_components, columns=[f"PC{i+1}" for i in range(n_components)])
```

9- Utilize SVM with the features obtained from step 8 in the PCA. Compare the results with the previous models that utilized the dropped correlated features.

10- Include the "%%time" command in each cell for modeling to measure the time required for creating each model, and report the duration for each model creation.

```python
%%time

X_train, X_test, y_train, y_test = train_test_split(df_pca, y, test_size=0.2, random_state=42)

svm_classifier = SVC()

svm_classifier.fit(X_train,y_train)

y_pred = svm_classifier.predict(X_test)

accuracy = accuracy_score(y_test, y_pred)
print("Accuracy:", accuracy)
```

In comparison with the output of the model with the dropped features, this model is more accurate and leverages better training at a significantly reduced CPU time. This model is also more accurate due to how it is able to leverage better datasets which allows for appropriate testing and refinement of accuracy, as learning models require redundancy to help tune and train the learning algorithms appropriately as these algorithms need clear and enforcing patterns to help guide it on what is truly correct; instead of providing unique but correct datasets that does not give much information to begin with. This also explains why this algorithm as a high degree of accuracy but is more efficient in comparison with the previous models involving the reduction of feature set.
