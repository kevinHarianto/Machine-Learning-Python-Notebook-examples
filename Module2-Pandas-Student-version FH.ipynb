# Python Dictionaries
Dictionaries are used to store data values in key:value pairs.

A dictionary is a collection which is ordered*, changeable and do not allow duplicates.


<br>

```python
thisdict = {
  "brand": "Ford",
  "model": "Mustang",
  "year": 1964
}
x = thisdict["model"] # get item
thisdict["year"]="2000" #update item
thisdict["color"] = "red" # Add item
del thisdict["model"] #Remove item
#access item
for x in thisdict.values():
    print(x)  #notice that we don't see the model anymore
```

# Series Data Structure

A Pandas Series is like a column in a table.

It is a one-dimensional array holding data of any type.

```python
import pandas as pd
```

You can also use a key/value object, like a dictionary, when creating a Series.

```python
d = {'a': 1, 'b': 2, 'c': 3}
ser = pd.Series(data=d, index=['a', 'b', 'c'])
ser
```

The keys of the dictionary become the Index values, hence the Index
values have no effect.

```python
d = {'a': 1, 'b': 2, 'c': 3}
ser = pd.Series(data=d, index=['x', 'y', 'z'])
ser
```

# Why we are using series instead of list or dictionary

In Python, the choice between using a list, dictionary, or a Pandas Series depends on the specific use case and the nature of the data you are working with. 

Structured Data: Series and DataFrames are designed to work with structured data. They provide more structure to your data compared to lists and dictionaries. 

Vectorized Operations:Series support vectorized operations, which can be more efficient than operations on lists or dictionaries.

Data Indexing: Series supports both integer-based and label-based indexing. This can make data manipulation easier.

Handling Missing Data: Series have built-in functionality for handling missing data. 

```python
#this creates a big series of random numbers
import numpy as np
s = pd.Series(np.random.randint(0,1000,10000))
s.head()
```
