#### <p style="color:red;">Exercise 5:

<p style="color:red;"> Find the minimun and maximum value of exploitability_score. Create a new column and name it "Exploitable", if base_score is above midrange, "Exploitable" is 1, otherwise it is 0. Use apply and lambda
Hint: The quantity obtained by adding the largest and smallest values and dividing by 2, statisticians call the midrange. 

```python
df['exploitability_score'].min()
```

```python
df['exploitability_score'].max()
```

```python
midrange = (df['exploitability_score'].min()+df['exploitability_score'].max())/2
df['Exploitable']=df.apply(lambda x: 1 if x['base_score'] > midrange else 0 ,  axis=1)
df['Exploitable']
```
